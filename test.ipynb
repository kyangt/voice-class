```python
import librosa

audio_path = './dataset/kaggle/kaggle-voice-data/one_sentence/S_01_4001_VE.wav'
data, sr = librosa.load(audio_path)

print(data.shape)
print(sr)
```

```python
spec = librosa.feature.melspectrogram(y=data, sr=sr)

import matplotlib.pyplot as plt
import librosa.display
librosa.display.specshow(spec, y_axis='mel', x_axis='s', sr=sr)
plt.colorbar()
```

```python
mfcc = librosa.feature.mfcc(y=data, sr=sr)

librosa.display.specshow(mfcc, x_axis='s', sr=sr)
plt.colorbar()
```
